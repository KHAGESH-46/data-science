```python
import pulp

# 1. Define the Problem
prob = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# 2. Define Decision Variables (Number of Product A and B to make)
A = pulp.LpVariable('Product_A', lowBound=0, cat='Integer')
B = pulp.LpVariable('Product_B', lowBound=0, cat='Integer')

# 3. Objective Function (Maximize Profit)
prob += 40 * A + 30 * B, "Total_Profit"

# 4. Constraints
prob += 2 * A + 1 * B <= 100, "Labor_Constraint"
prob += A + B <= 40, "Material_Constraint"

# 5. Solve
prob.solve()

# 6. Output Insights
print(f"Status: {pulp.LpStatus[prob.status]}")
print(f"Produce {A.varValue} units of Product A")
print(f"Produce {B.varValue} units of Product B")
print(f"Maximum Profit: ${pulp.value(prob.objective)}")
```
